```python
G = np.ones_like(A,dtype='f')
for i in range(len(A)):
    G[i]=20*math.log(A[i]/max(A),10)
plt.style.use('_mpl-gallery')
fig, ax = plt.subplots()
ax.plot(f,G)
ax.set(xlim=(38.47,42.37),
       ylim=(-18,0.5))
plt.title("Gain en fonction de la fréquence")
plt.xlabel("Fréquence en kHz")
plt.ylabel ("Gain en dB")
plt.yticks(np.arange(-21,0,3))
plt.show()
```
